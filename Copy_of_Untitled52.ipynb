```python
#remove all duplicates value

df.drop_duplicates(inplace=True)
```

```python
df['City'].fillna(df['City'].mode()[0],inplace=True)
```

```python
df['Sales'].fillna(df['Sales'].median(),inplace=True)
```

```python
df['Discount'].fillna(df['Discount'].median(),inplace=True)
```

```python
df['Profit'].fillna(df['Profit'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['City'].unique()
```

```python
df['City']=df['Discount']*df['Profit']
```

```python
df.head()
```

```python
#converts dates
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
#converts dates
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df.head()
```

```python
# extract month and year
df['Order Month']=df['Order Date'].dt.month
df['Order Year']=df['Order Date'].dt.year
```

```python
df.head()
```

```python
# profit margin
df['Profit Margin']=df['Profit']/df['Sales']
```

```python
df['Sales Category']=pd.cut(df['Sales'],
                            bins=[0,100,500,1000,5000],
                            labels=['Low','Medium','High','Very High'])
```

```python
df.head()
```

```python
#  which category is the highest  total sales

df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
# which category is the most profitable
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```
